```python
import psycopg2
import time
# PostgreSQL 연결 설정
DB_CONFIG = {
    "dbname": "book_recommend",
    "user": "postgres",
    "password": "1234",
    "host": "localhost",
    "port": "5432",
}

# 더미 데이터 리스트
dummy_data = [
    (1, 101, 1, 1, True, "2025-02-03T12:00:00"),
    (2, 101, 1, 2, False, "2025-02-03T12:05:00"),
    (3, 101, 2, 3, True, "2025-02-03T12:10:00"),
    (4, 101, 2, 4, False, "2025-02-03T12:15:00"),
    (5, 101, 3, 5, True, "2025-02-03T12:20:00"),
    (6, 101, 3, 6, False, "2025-02-03T12:25:00"),
    (7, 101, 4, 7, True, "2025-02-03T12:30:00"),
    (8, 101, 4, 8, False, "2025-02-03T12:35:00"),
    (9, 101, 5, 9, True, "2025-02-03T12:40:00"),
    (10, 101, 5, 10, False, "2025-02-03T12:45:00"),
    (11, 101, 6, 11, True, "2025-02-03T12:50:00"),
    (12, 101, 6, 12, False, "2025-02-03T12:55:00"),
    (13, 101, 7, 13, True, "2025-02-03T13:00:00"),
    (14, 101, 7, 14, False, "2025-02-03T13:05:00"),
    (15, 101, 8, 15, True, "2025-02-03T13:10:00"),
    (16, 101, 8, 16, False, "2025-02-03T13:15:00"),
    (17, 101, 9, 17, True, "2025-02-03T13:20:00"),
    (18, 101, 9, 18, False, "2025-02-03T13:25:00"),
    (19, 101, 10, 19, True, "2025-02-03T13:30:00"),
    (20, 101, 10, 20, False, "2025-02-03T13:35:00"),
    (21, 102, 1, 21, True, "2025-02-03T13:40:00"),
    (22, 102, 1, 22, False, "2025-02-03T13:45:00"),
    (23, 102, 2, 23, True, "2025-02-03T13:50:00"),
    (24, 102, 2, 24, False, "2025-02-03T13:55:00"),
    (25, 102, 3, 25, True, "2025-02-03T14:00:00"),
    (26, 102, 3, 26, False, "2025-02-03T14:05:00"),
    (27, 102, 4, 27, True, "2025-02-03T14:10:00"),
    (28, 102, 4, 28, False, "2025-02-03T14:15:00"),
    (29, 102, 5, 29, True, "2025-02-03T14:20:00"),
    (30, 102, 5, 30, False, "2025-02-03T14:25:00"),
]

# 테이블명
TABLE_NAME = "user_responses"

def insert_dummy_data():
    """더미 데이터를 한 회차씩 DB에 삽입하는 함수"""
    conn = psycopg2.connect(**DB_CONFIG)
    cur = conn.cursor()

    # 삽입된 question_number를 추적
    inserted_question_numbers = set()

    try:
        for row in dummy_data:
            id, user_id, question_number, sentence_id, is_positive, datetime = row
            
            # 이미 해당 question_number가 삽입되었으면 건너뛴다.
            if question_number in inserted_question_numbers:
                continue

            # 데이터 삽입
            cur.execute(
                f"""
                INSERT INTO {TABLE_NAME} (id, user_id, question_number, sentence_id, is_positive, datetime)
                VALUES (%s, %s, %s, %s, %s, %s)
                """,
                (id, user_id, question_number, sentence_id, is_positive, datetime)
            )

            conn.commit()
            inserted_question_numbers.add(question_number)

            print(f"✅ Inserted question_number {question_number} data")

            # 5초 대기 후 다음 회차 데이터 삽입
            time.sleep(5)

    except Exception as e:
        print("❌ Error:", e)
        conn.rollback()
    finally:
        cur.close()
        conn.close()

if __name__ == "__main__":
    insert_dummy_data()
```
