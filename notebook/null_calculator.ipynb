# 수집한 데이터 중 None 값 확인

```python
import json
import re

# JSON 데이터 로드
with open('/Users/hyo-cheolahn/Documents/Projects/Book-Recommendation/data/scraping/all_book_data_ver3.json', "r", encoding="utf-8") as file:
    data = json.load(file)

print(len(data))
# description과 key_sentences를 통합한 리스트 생성
title_loss_idx = []
description_loss_idx = []
key_sentences_loss_idx = []

description_list = []
key_sentences_list = []
isbn_list = []
isbn_loss_idx = []
author_list = []
author_loss_idx = []
publisher_list = []
publisher_loss_idx = []
img_url_list = []
img_url_loss_idx = []
publication_date_list = []
publication_date_loss_idx = []

for idx, el in enumerate(data):
    # description과 key_sentences를 가져옴
    title = el.get('title', None)
    description = el.get('description', None)  # description이 없으면 빈 문자열
    key_sentences = el.get('key_sentences', None)  # key_sentences가 없으면 빈 문자열
    isbn_str = str(el.get('isbn13', ''))  # isbn13을 문자열로 변환
    author = el.get('author', None)
    publisher = el.get('publisher', None)
    img_url = el.get('img_url', None)
    publication_date = el.get('publication_date', None)
    
    


    # 숫자만 추출하고 빈 문자열인 경우 처리
    isbn_numeric = re.sub(r"\D", "", isbn_str)
    if isbn_numeric:  # 숫자가 있는 경우에만 int 변환
        isbn13 = int(isbn_numeric)
    else:
        isbn13 = None
        isbn_loss_idx.append(idx)

    if not title:
        title_loss_idx.append(idx)
    if not description:
        description_loss_idx.append(idx)
    if not key_sentences:
        key_sentences_loss_idx.append(idx)
    if not author:
        author_loss_idx.append(idx)
    if not publisher:
        publisher_loss_idx.append(idx)
    if not img_url:
        img_url_loss_idx.append(idx)
    if publication_date is None:
        publication_date_loss_idx.append(idx)

    # 리스트에 추가
    description_list.append(description)
    key_sentences_list.append(key_sentences)
    isbn_list.append(isbn13)
    author_list.append(author)
    publisher_list.append(publisher)
    img_url_list.append(img_url)
    publication_date_list.append(publication_date)

# 결과 확인
print(f"총 {len(key_sentences_list)}개의 항목이 생성되었습니다.")
```
