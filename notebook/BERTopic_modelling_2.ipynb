### 불용어 처리 추가

```python
# 불용어 리스트 정의
with open('stopwords-ko.txt', 'r', encoding='utf-8') as file:
    stopwords = [line.strip() for line in file]


# 형태소 분석 및 불용어 제거
filtered_df = []  # 문서별로 필터링된 단어 저장
for line in filtered_df:
    filtered_words = []  # 현재 문서의 단어 저장
    parsed = mecab.parse(line).splitlines()  # 형태소 분석 결과를 라인별로 분리
    for token in parsed:
        if not token or token == "EOS":  # 빈 토큰 처리 및 EOS 제외
            continue
        word = token.split('\t')[0]  # 단어 추출
        if word not in stopwords:  # 불용어 제거
            filtered_words.append(word)
    filtered_df.append(filtered_words)

# 결과 출력
# for idx, words in enumerate(filtered_words_per_doc):
#     print(f"문서 {idx + 1}: {words}")
```

```python
if filtered_df.columns.size == 1:
    documents = filtered_df.iloc[:, 0].tolist()
len(documents)

preprocessed_documents = []

for line in tqdm(documents):
  # 빈 문자열이거나 숫자로만 이루어진 줄은 제외
  if line and not line.replace(' ', '').isdecimal():
    preprocessed_documents.append(line)
```

```python

class CustomTokenizer:
    def __init__(self, tagger):
        self.tagger = tagger
    def __call__(self, sent):
        word_tokens = self.tagger.morphs(sent)
        result = [word for word in word_tokens if len(word) > 1]
        return result
```
